## Grover’s Algorithm Setup (3-Qubit System)

- **Marked element**: $\ket{a} = \ket{101}$, which corresponds to decimal $a = 5$

- **Initial state**:  
  Apply Hadamard to all qubits starting from $\ket{000}$.  
  This gives the uniform superposition:
  $
  \ket{\phi} = \frac{1}{\sqrt{8}} \sum_{x=0}^{7} \ket{x}
  = \frac{1}{\sqrt{8}}(\ket{000} + \ket{001} + \cdots + \ket{111})
  $

- **Oracle operator**:
  $
  V = I - 2\ket{a}\bra{a}
  $

- **Diffusion operator**:
  $
  W = 2\ket{\phi}\bra{\phi} - I
  $


## Understanding Inner Product and Outer Product

- **Inner product** $\bra{a}\ket{\phi}$:  
  A scalar that tells **how much** $\ket{\phi}$ "overlaps" with $\ket{a}$  
  In our case:
  $\bra{a}\ket{\phi} = \frac{1}{\sqrt{8}}$

  **Exercise** Prove $\bra{a}\ket{\phi} = \frac{1}{\sqrt{8}}$.

- **Outer product** $\ket{a}\bra{a}$:  
  This is an **operator**, also called a **projector**. It is a **projector** that projects any input state onto the direction of $\ket{a}$.
  It acts on a state to **extract the component in the direction of** $\ket{a}$. 
  So if we apply this to any $\ket{\psi}$, we get: $\ket{a}\bra{a} \ket{\psi} = \braket{a|\psi} \cdot \ket{a}$.
  This means:
    - Take the **inner product** between $\ket{a}$ and $\ket{\psi}$ (a scalar),
    - Then **scale $\ket{a}$** by that scalar.

- What is $\ket{a}\bra{a}\ket{\phi}$?
  This is the application of the projector $\ket{a}\bra{a}$ to the superposition state $\ket{\phi}$.
  It equals a **scaled version** of $\ket{a}$, where the scale is the **amplitude of $\ket{a}$ in $\ket{\phi}$**.
  1. Compute the inner product:$\braket{a|\phi} = \frac{1}{\sqrt{8}}$
  2. Multiply that scalar with $\ket{a}$: $\ket{a}\bra{a}\ket{\phi} = \frac{1}{\sqrt{8}} \cdot \ket{a} = \frac{1}{\sqrt{8}} \ket{101}$


## First Iteration

### Step 01: 

Apply the oracle to the superposition:
$
V\ket{\phi} = (I - 2\ket{a}\bra{a}) \ket{\phi}
$

#### Set up:

- Marked state:  $\ket{a} = \ket{101}$
- Oracle operator:  $V = I - 2\ket{a}\bra{a} = I - 2\ket{101}\bra{101}$

- Uniform superposition (after Hadamards):  
  $
  \ket{\phi} = \frac{1}{\sqrt{8}} \sum_{x=0}^{7} \ket{x}
  = \frac{1}{\sqrt{8}} (\ket{000} + \ket{001} + \ket{010} + \ket{011} + \ket{100} + \ket{101} + \ket{110} + \ket{111})
  $


#### Step 1: Expand the expression

$
V\ket{\phi} = \ket{\phi} - 2\ket{a}\bra{a}\ket{\phi}
$


#### Step 2: Compute the projection

From earlier:
$
\bra{a}\ket{\phi} = \frac{1}{\sqrt{8}} \quad \Rightarrow \quad
\ket{a}\bra{a}\ket{\phi} = \frac{1}{\sqrt{8}} \ket{a} = \frac{1}{\sqrt{8}} \ket{101}
$


#### Step 3: Multiply by 2

$
2\ket{a}\bra{a}\ket{\phi} = \frac{2}{\sqrt{8}} \ket{101}
$


#### Step 4: Subtract from $\ket{\phi}$

Recall:
$
\ket{\phi} = \frac{1}{\sqrt{8}} ( \ket{000} + \ket{001} + \ket{010} + \ket{011} + \ket{100} + \ket{101} + \ket{110} + \ket{111} )
$

Subtract the scaled projection:
$
V\ket{\phi} = \ket{\phi} - \frac{2}{\sqrt{8}} \ket{101}
$

So the amplitude of $\ket{101}$ becomes:
$
\frac{1}{\sqrt{8}} - \frac{2}{\sqrt{8}} = -\frac{1}{\sqrt{8}}
$

#### Final Result

$
V\ket{\phi} = \frac{1}{\sqrt{8}} (
\ket{000} + \ket{001} + \ket{010} + \ket{011} + \ket{100} - \ket{101} + \ket{110} + \ket{111}
)
$

The oracle **flips the sign** of the amplitude of the marked state $\ket{101}$.

### Step 02: Diffusion 

Apply $W$ to $V\ket{\phi}$

- The diffusion operator is:
  $W = 2\ket{\phi}\bra{\phi} - I$

- From the previous oracle step, we have:
  $V\ket{\phi} = \frac{1}{\sqrt{8}} (\ket{000} + \ket{001} + \ket{010} + \ket{011} + \ket{100} - \ket{101} + \ket{110} + \ket{111})$


#### Goal

Apply the diffusion operator to the post-oracle state:

$W V \ket{\phi} = (2\ket{\phi}\bra{\phi} - I) V\ket{\phi}$

We'll do this in steps.

#### Step 1: Expand the formula

$W V \ket{\phi} = 2 \braket{\phi | \psi} \cdot \ket{\phi} - \ket{\psi}$

Here, $\ket{\psi} = V\ket{\phi}$

#### Step 2: Compute $\braket{\phi|\psi}$

Recall:

- $\ket{\phi} = \frac{1}{\sqrt{8}} \sum_{x=0}^{7} \ket{x}$
- $\ket{\psi} = V\ket{\phi} = \frac{1}{\sqrt{8}} (\text{all basis states, but with } \ket{101} \text{ having a negative sign})$

So we compute:

$\braket{\phi|\psi} = \frac{1}{8} (1 + 1 + 1 + 1 + 1 - 1 + 1 + 1) = \frac{6}{8} = \frac{3}{4}$

#### Step 3: Multiply $2 \braket{\phi|\psi} \cdot \ket{\phi}$

This gives:

$2 \cdot \frac{3}{4} \cdot \ket{\phi} = \frac{3}{2} \cdot \ket{\phi}$

#### Step 4: Subtract $\ket{\psi}$

Recall:

$\ket{\phi} = \frac{1}{\sqrt{8}} \sum_{x=0}^{7} \ket{x}$

$\ket{\psi} = \frac{1}{\sqrt{8}} (\ket{000} + \ket{001} + \ket{010} + \ket{011} + \ket{100} - \ket{101} + \ket{110} + \ket{111})$

Now compute:

$W\ket{\psi} = \frac{3}{2} \cdot \ket{\phi} - \ket{\psi}$

This affects only the amplitudes:

- For all $x \neq 101$: amplitude is $\frac{3}{2\sqrt{8}} - \frac{1}{\sqrt{8}} = \frac{1}{2\sqrt{8}}$
- For $x = 101$: amplitude is $\frac{3}{2\sqrt{8}} - (-\frac{1}{\sqrt{8}}) = \frac{5}{2\sqrt{8}}$

#### Final Result

$W V \ket{\phi} = \frac{1}{2\sqrt{8}} \sum_{\substack{x=0 \\ x \neq 5}}^7 \ket{x} + \frac{5}{2\sqrt{8}} \ket{101}$

This means:
- The amplitude of the marked state $\ket{101}$ is now significantly amplified
- The other states have lower amplitude

This is the core mechanism of amplitude amplification in Grover's algorithm.

# Grover's algorithm: A different narrative (3-qubits setup)


We begin with $3$-qubits, all in $\ket{0}^{\otimes 3} = \ket{000}$. This means every qubit is in state $\ket{0}$. 
Then, we apply Hadamard gates to all qubits. This will transform $H^{\otimes 3}$ to this state: $\ket{\phi} = H^{\otimes 3} \ket{0}^{\otimes 3}.$

The Hadamard gate acts as: $H\ket{0} = \frac{1}{\sqrt{2}} (\ket{0} + \ket{1}).$ So applying it to each qubit gives a uniform superposition of all possible states:
$$
\ket{\phi} = \frac{1}{\sqrt{2^3}} \sum_{x \in \{0,1\}^3} \ket{x} = \frac{1}{\sqrt{8}} (\ket{000} + \ket{001} + \ket{010} + \ket{011} + \ket{100} + \ket{101} + \ket{110} + \ket{111})
$$


This is the key definition of $\ket{\phi}$ — a balanced superposition of all possible inputs.

The algorithm assumes one of these states (call it $\ket{a}=\ket{101}$) satisfies $f(x)=1$.
We can then split the superposition into two parts:
- the marked state $\ket{a}$, and
- all the other states (call their uniform combination $\ket{r}$).

So: $\ket{\phi} = \frac{1}{\sqrt{8}}\ket{a} + \frac{1}{\sqrt{8}} \sum_{i \neq a} \ket{i}$





Now we define:
$\ket{r} = \frac{1}{\sqrt{8-1}} \sum_{i \neq a} \ket{i}.$

This allows us to rewrite
$\ket{\phi} = \frac{1}{\sqrt{8}}\ket{a} + \sqrt{\frac{8-1}{8}} \ket{r}.$

Here, $\ket{\phi}$ is written in terms of two orthogonal states, $\ket{a}$ and $\ket{r}$. From linear algebra and trigonometry, we know that any vector expressed as a sum of two perpendicular components can be represented using an angle $\theta$ that describes how much the vector “leans” toward one direction or the other.
Therefore, we can rewrite $\ket{\phi}$ in the form

$\ket{\phi} = \cos\theta \ket{r} + \sin\theta  \ket{a}$

where the coefficients $\cos\theta$ and $\sin\theta$ correspond to the projections of $\ket{\phi}$ onto the directions of $\ket{r}$ and $\ket{a}$, respectively.
We should remind here that every valid quantum state (here $\ket{\phi}$) has a norm equal to one.


Considering $\ket{\phi} = \frac{1}{\sqrt{8}}\ket{a} + \sqrt{\frac{8-1}{8}} \ket{r}$ we can define $\theta$ such that:
$$
\sin \theta = \frac{1}{\sqrt{8}}, \quad \cos \theta = \sqrt{\frac{8-1}{8}}.
$$

This expression shows that geometrically in the 2D plane $\ket{\phi}$ “leans” slightly toward $\ket{r}$ (the non-solution space), since $\cos\theta$ is almost 1 for large $N$. It’s **initially** very close to $\ket{r}$ and only slightly toward $\ket{a}$.

In order to find the solution, i.e. $\ket{a}$, by applying some operations in several iterations we need to aligne $\ket{\phi}$ more towards $\ket{a}$. This is basically what Grover's algorithm does. In each iteration, by applying **reflections** the algorithm rotates $\ket{\phi}$ toward $\ket{a}$ by an angle $2 \theta$, so after roughly $\frac{\pi}{4}\sqrt{N}$ iterations, it points very close to $\ket{a}$. Let's try this in the following.

Grover's algorithm in general employs two operations:
- The Oracle Operation
- The Diffusion Operation

### The Oracle Operation

The oracle is a quantum operator that recognizes the marked state $\ket{a}$.
It flips the phase of $\ket{a}$ while leaving all other states unchanged:

$$
V \ket{x} =
\begin{cases}
    - \ket{x}, & \text{if } x = a, \
\\ \ket{x}, & \text{otherwise.}
\end{cases}
$$

This can be written as:

$$
V = I - 2 \ket{a}\bra{a}.
$$

When we apply this to $\ket{\phi}$, the amplitude of $\ket{a}$ changes sign, i.e. it is reflected across the subspace orthogonal to $\ket{a}$:

$
\ket{\phi_1} = V \ket{\phi} = U_f ( \frac{1}{\sqrt{8}} (\ket{000} + \ket{001} + \ket{010} + \ket{011} + \ket{100} + \ket{101} + \ket{110} + \ket{111}) ) = \frac{1}{\sqrt{8}} (\ket{000} + \ket{001} + \ket{010} + \ket{011} + \ket{100} - \ket{101} + \ket{110} + \ket{111})
$

**Exercise** Show computations steps of $V \ket{\phi}$.


In our 2D geometric picture (spanned by ${\ket{a}, \ket{r}}$), the oracle reflection flips $\ket{\phi}$ over the $\ket{r}$ axis.
So, after applying the oracle, the state becomes:

$$
\ket{\phi_1} = V \ket{\phi} = \cos\theta \ket{r} - \sin\theta \ket{a}.
$$


### The Diffusion Operator (Inversion about the Mean)

[some text about intuition behind this in simple language]
Now, we need to “invert about the mean” — this is called the Grover diffusion operator, defined as:

$$
W = 2\ket{\phi}\bra{\phi} - I.
$$

This operator reflects the current state ($\ket{\phi_1}$) around $\ket{\phi}$ itself.
Geometrically, in the 2D plane of ${\ket{a}, \ket{r}}$, this reflection flips the state across $\ket{\phi}$, effectively rotating it toward $\ket{a}$ by $2\theta$.

Applying it to $\ket{\phi_1}$ gives:

$$
\ket{\phi_2} = W \ket{\phi_1}.
$$

Step 3. One Full Grover Iteration

Let’s summarize what happens in one iteration:
- Oracle: reflection over $\ket{r}$ flips the phase of $\ket{a}$ component.

$\ket{\phi} \mapsto \cos\theta \ket{r} - \sin\theta \ket{a}$.

- Diffusion: reflection about $\ket{\phi}$ rotates the state by $2\theta$ toward $\ket{a}$.

$\ket{\phi’} \mapsto \cos(3\theta)\ket{r} + \sin(3\theta)\ket{a}$.

After $k$ iterations, the state becomes:

$$
\ket{\psi_k} = \cos((2k+1)\theta)\ket{r} + \sin((2k+1)\theta)\ket{a}.
$$

So each iteration increases the amplitude of $\ket{a}$ by roughly $2\theta$.

⸻

Step 4. How Many Iterations?

To maximize the probability of measuring $\ket{a}$, we want the state to align as closely as possible with $\ket{a}$.
That happens when:

$$
(2k+1)\theta \approx \frac{\pi}{2}.
$$

Solving for $k$ gives:

$$
k \approx \frac{\pi}{4\theta} - \frac{1}{2}.
$$

For our case with $N=8$:

$$
\sin\theta = \frac{1}{\sqrt{8}} \Rightarrow \theta \approx 0.36~\text{radians}.
$$

Then:

$$
k \approx \frac{\pi}{4(0.36)} - \frac{1}{2} \approx 2.68.
$$

So we perform 2 to 3 iterations to reach near-perfect alignment with $\ket{a}$.

⸻

Step 5. Measurement and Outcome

After those iterations, the state $\ket{\psi_k}$ is very close to $\ket{a}$:

$$
\ket{\psi_k} \approx \ket{a}.
$$

When we measure the qubits in the computational basis, we will get $\ket{a} = \ket{101}$ with high probability.

# Grover's algorithm: General


We start with all qubits in the zero state

We begin with $n$ qubits, all in $\ket{0}^{\otimes n} = \ket{0 0 \dots 0}$. This means every qubit is in state $\ket{0}$.

Then, we apply Hadamard gates to all qubits. This will transform $H^{\otimes n}$ to this state: $\ket{\phi} = H^{\otimes n} \ket{0}^{\otimes n}.$

The Hadamard gate acts as: $H\ket{0} = \frac{1}{\sqrt{2}} (\ket{0} + \ket{1}).$

So applying it to each qubit gives a uniform superposition of all possible $n$-bit states:
$$
\ket{\phi} = \frac{1}{\sqrt{N}} \sum_{x \in \{0,1\}^n} \ket{x},
$$
where $N = 2^n$ is the number of basis states.

This is the key definition of $\ket{\phi}$ — a balanced superposition of all possible inputs.

The algorithm assumes one of these states (call it $\ket{a}$) satisfies $f(x)=1$.
We can then split the superposition into two parts:
- the marked state $\ket{a}$, and
- all the other states (call their uniform combination $\ket{r}$).

So: $\ket{\phi} = \frac{1}{\sqrt{N}}\ket{a} + \frac{1}{\sqrt{N}} \sum_{i \neq a} \ket{i}$

Now we define:
$$
\ket{r} = \frac{1}{\sqrt{N-1}} \sum_{i \neq a} \ket{i}.
$$

This allows us to rewrite $\ket{\phi}$ as:
$$
\ket{\phi} = \frac{1}{\sqrt{N}}\ket{a} + \sqrt{\frac{N-1}{N}} \ket{r}.
$$

⸻

1. Expressing $\ket{\phi}$ using an angle $\theta$

Let’s define an angle $\theta$ such that:
$$
\sin \theta = \frac{1}{\sqrt{N}}, \quad \cos \theta = \sqrt{\frac{N-1}{N}}.
$$

Then we can write:
$$
\ket{\phi} = \sin\theta \ket{a} + \cos\theta \ket{r}.
$$

This expression shows that $\ket{\phi}$ “leans” slightly toward $\ket{r}$ (the non-solution space), since $\cos\theta$ is almost 1 for large $N$.

⸻

5. Geometric interpretation

In Grover’s algorithm, $\ket{\phi}$ lies in the 2D plane spanned by $\ket{a}$ and $\ket{r}$.
It’s initially very close to $\ket{r}$ and only slightly toward $\ket{a}$.

Each Grover iteration (two reflections) rotates $\ket{\phi}$ toward $\ket{a}$ by an angle $2\theta$, so after roughly $\frac{\pi}{4}\sqrt{N}$ iterations, it points very close to $\ket{w}$.
